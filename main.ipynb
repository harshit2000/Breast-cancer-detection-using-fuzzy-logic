### Cluster Analysis for breaking into sets

```python
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
df.columns
```

```python
# Clustering into relevant cluster of low medium or high values
for i in df.columns:
    if i == 'id' or i == 'class':
        continue
    if i == 'clump_thickness' or i == 'normal_nucleoli' :
        # 3 clusters for clump_thickness and normal_nucleoli
        kmeans = KMeans(n_clusters=3)
    else:
        # 2 clusters for the rest
        kmeans = KMeans(n_clusters=2)
    
    X = np.asarray(df[i]).reshape((-1,1))
    y = np.asarray(df['class']).reshape((-1,1))
    kmeans.fit(X,y)
    pred = kmeans.predict(X).reshape((-1,1))
    data = np.hstack((X,pred))
    data = pd.DataFrame(data)
    data.columns = [i,'predClass']
    #data.head()
    sns.countplot(i, data = data, hue='predClass')
    plt.show()
```

### Train Test Split

```python
df = pd.read_csv('Categorical_data(Fuzzy Sets).csv')
```

```python
# 70 % train-test split
split = int(0.7*df.shape[0])
train_data = df[:split]
test_data = df[split:]
test_data = test_data.reset_index(drop=True)
```

```python
train_data.shape, test_data.shape
```

```python
df.columns
```

```python
input_cols = ['clump_thickness', 'uniformity_cell_size',
       'uniformity_cell_shape', 'marginal_adhesion', 'single_epi_cell_size',
       'bare_nuclei', 'bland_chromatin', 'normal_nucleoli', 'mitosis']
output_cols = ['class']
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
sk_tree = DecisionTreeClassifier(criterion='entropy', max_depth=5)
```

```python
X_train = np.asarray(train_data[input_cols])
y_train = np.asarray(train_data[output_cols])
y_train=y_train.astype('int')

X_test = np.asarray(test_data[input_cols])
y_test = np.asarray(test_data[output_cols])
y_test=y_test.astype('int')
```

```python
sk_tree.fit(X_train, y_train)
```

```python
pred = sk_tree.predict(X_test)
```

```python
pred.shape
```

```python
# training accuracy
sk_tree.score(X_train, y_train)
```

```python
sk_tree.score(X_test, y_test)
```
