```python
pred = []
for i in range(X.shape[0]):
    val = eng.evalFuzzy(X[i][0],X[i][1],X[i][2],X[i][3],X[i][4],X[i][5],X[i][6],X[i][7],X[i][8])
    pred.append(val)
```

```python
def accuracy(pred, y ,split):
    y_calc = []
    right = 0
    wrong = 0
    for i in range(pred.shape[0]):
        if pred[i] < split:
            y_calc.append(2)
        else:
            y_calc.append(4)
    
    y_calc = np.array(y_calc)
    
    for i in range(y_calc.shape[0]):
        if y_calc[i] == y[i]:
            right+=1
        else:
            wrong+=1
            
    return right/y.shape[0]
    
```

```python
split_arr = (np.arange(50)+50)/100
```

```python
split_arr
```

```python
acc = []
for i in split_arr:
    acc.append(accuracy(pred,y,i))
```
